```python
import matplotlib.pyplot as plt
import pickle as pkl
from scipy.stats import gaussian_kde
import numpy as np
import math
import os
```

## bias boxplots

### A

```python
n_obs_list = [100, 500, 1000, 5000]
simulation_functions = [
    lambda n: n,
    lambda n: int(n * math.log(n)),
    lambda n: int(n ** (3/2)),
    lambda n: n ** 2
]
n_sims_label = [r'$n$', r'$n \log(n)$', r'$n^{3/2}$', r'$n^2$']

num_seeds = 101
true_param = 3.0

base_dir = "../res/gnk/"

for n_obs in n_obs_list:
    all_mean_biases = []
    labels = []
    for ii, sim_fn in enumerate(simulation_functions):
        n_sims = sim_fn(n_obs)
        mean_biases = []
        for i in range(num_seeds):
            dir_path = os.path.join(base_dir, f"npe_n_obs_{n_obs}_n_sims_{n_sims}_seed_{str(i)}")
            if os.path.exists(dir_path):
                full_path = os.path.join(dir_path, "posterior_samples.pkl")
                try:
                    with open(full_path, 'rb') as file:
                        posterior_samples = pkl.load(file)
                        seed_mean_bias = np.mean(posterior_samples, axis=0)[0] - true_param
                        mean_biases.append(seed_mean_bias)
                except FileNotFoundError:
                    continue
        all_mean_biases.append(mean_biases)
        labels.append(n_sims_label[ii])

    plt.figure(figsize=(10, 6))
    plt.boxplot(all_mean_biases, labels=labels, patch_artist=True,
                boxprops=dict(facecolor='lightblue', color='blue'),
                medianprops=dict(color='red', linewidth=2),
                whiskerprops=dict(color='blue'), capprops=dict(color='blue'))
    # plt.title(f'Boxplot of Mean Biases for $n_{{\text{{obs}}}}={n_obs}$', fontsize=16)
    plt.xlabel('Number of Simulations', fontsize=24)
    plt.ylabel('Mean Bias', fontsize=24)
    plt.axhline(y=0, color='black', linestyle='dashed', linewidth=1.5)
    # plt.legend(fontsize=12)
    plt.tick_params(axis='both', which='major', labelsize=18)
    # plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.savefig(f'plots/gnk_A_bias_boxplot_n_obs_{n_obs}.pdf')
    plt.close()
```

## octiles vs duodeciles

```python
import os
import math
import numpy as np
import matplotlib.pyplot as plt

n_obs_list = [100, 500, 1000, 5000]
simulation_functions = [
    lambda n: n,
    lambda n: int(n * math.log(n)),
    lambda n: int(n ** (3/2)),
    lambda n: n ** 2
]
n_sims_label = [r'$n$', r'$n \log(n)$', r'$n^{3/2}$', r'$n^2$']

num_seeds = 101

base_dir_octiles = "../res/gnk/"
base_dir_duodeciles = "../res/gnk_duodeciles/"

for n_obs in n_obs_list:
    # Lists to store statistics for Octiles
    octile_medians = []
    octile_lower_errors = []
    octile_upper_errors = []
    
    # Lists to store statistics for Duodeciles
    duodecile_medians = []
    duodecile_lower_errors = []
    duodecile_upper_errors = []
    
    for ii, sim_fn in enumerate(simulation_functions):
        n_sims = sim_fn(n_obs)
        kl_values_octile = []
        kl_values_duodecile = []
        
        for i in range(num_seeds):
            # Octiles
            dir_path_octile = os.path.join(
                base_dir_octiles,
                f"npe_n_obs_{n_obs}_n_sims_{n_sims}_seed_{str(i)}"
            )
            kl_file_octile = os.path.join(dir_path_octile, "kl.txt")
            if os.path.isfile(kl_file_octile):
                with open(kl_file_octile, 'r') as file:
                    try:
                        kl_value = float(file.read().strip())
                        if math.isfinite(kl_value):
                            kl_values_octile.append(kl_value)
                    except ValueError:
                        continue
            # Duodeciles
            dir_path_duodecile = os.path.join(
                base_dir_duodeciles,
                f"npe_n_obs_{n_obs}_n_sims_{n_sims}_seed_{str(i)}"
            )
            kl_file_duodecile = os.path.join(dir_path_duodecile, "kl.txt")
            if os.path.isfile(kl_file_duodecile):
                with open(kl_file_duodecile, 'r') as file:
                    try:
                        kl_value = float(file.read().strip())
                        if math.isfinite(kl_value):
                            kl_values_duodecile.append(kl_value)
                    except ValueError as e:
                        continue
            else:
                dir_path_duodecile = os.path.join(
                    base_dir_duodeciles,
                    f"npe_n_obs_{n_obs}_n_sims_25000000_seed_{str(i)}"
                )
                print(dir_path_duodecile)
                if os.path.isfile(kl_file_duodecile):
                    with open(kl_file_duodecile, 'r') as file:
                        try:
                            kl_value = float(file.read().strip())
                            if math.isfinite(kl_value):
                                kl_values_duodecile.append(kl_value)
                        except ValueError as e:
                            print(e)
                            continue

        # Calculate statistics for Octiles
        print("kl_values_octile", kl_values_octile)
        if kl_values_octile:
            median_octile = np.median(kl_values_octile)
            lower_octile = np.percentile(kl_values_octile, 5)
            upper_octile = np.percentile(kl_values_octile, 95)
            octile_medians.append(median_octile)
            octile_lower_errors.append(median_octile - lower_octile)
            octile_upper_errors.append(upper_octile - median_octile)
        else:
            octile_medians.append(np.nan)
            octile_lower_errors.append(0)
            octile_upper_errors.append(0)
        
        # Calculate statistics for Duodeciles
        print("kl_values_duodecile", kl_values_duodecile)
        if kl_values_duodecile:
            median_duodecile = np.median(kl_values_duodecile)
            lower_duodecile = np.percentile(kl_values_duodecile, 5)
            upper_duodecile = np.percentile(kl_values_duodecile, 95)
            duodecile_medians.append(median_duodecile)
            duodecile_lower_errors.append(median_duodecile - lower_duodecile)
            duodecile_upper_errors.append(upper_duodecile - median_duodecile)
        else:
            print("No data for Duodeciles")
            duodecile_medians.append(np.nan)
            duodecile_lower_errors.append(0)
            duodecile_upper_errors.append(0)
    
    # Convert labels to x positions
    x_positions = np.arange(len(n_sims_label))
    
    # Offset for side-by-side plotting
    offset = 0.01  # Adjust as needed
    width = 0.05  # Width of the markers

    # Plotting
    plt.figure(figsize=(12, 8))
    
    # Octiles
    plt.errorbar(
        x_positions - offset,
        octile_medians,
        yerr=[octile_lower_errors, octile_upper_errors],
        fmt='-o',
        label='Octiles',
        capsize=5,
        markersize=8,
        linewidth=2
    )
    
    # Duodeciles
    plt.errorbar(
        x_positions + offset,
        duodecile_medians,
        yerr=[duodecile_lower_errors, duodecile_upper_errors],
        fmt='-s',
        linestyle='dashed',
        label='Duodeciles',
        capsize=5,
        markersize=8,
        linewidth=2
    )
    print("yes")
    plt.xticks(x_positions, n_sims_label, fontsize=24)
    plt.xlabel("Choice of N", fontsize=24)
    plt.ylabel(r"KL Divergence", fontsize=24)
    # plt.title(f"KL Divergence with Error Bars (n_obs = {n_obs})", fontsize=26)
    plt.legend(fontsize=20)
    plt.tick_params(axis='both', which='major', labelsize=24)
    plt.tight_layout()
    plt.savefig(f"plots/gnk_summaries_{str(n_obs)}.pdf")
    plt.clf()
```

```python
import os
import math
import numpy as np
import matplotlib.pyplot as plt

n_obs_list = [100, 500, 1000, 5000]
simulation_functions = [
    lambda n: n,
    lambda n: int(n * math.log(n)),
    lambda n: int(n ** (3/2)),
    lambda n: n ** 2
]
n_sims_label = [r'$n$', r'$n \log(n)$', r'$n^{3/2}$', r'$n^2$']

num_seeds = 101

base_dir_octiles = "../res/gnk/"
base_dir_duodeciles = "../res/gnk_duodeciles/"

for n_obs in n_obs_list:
    # Lists to store statistics of differences
    diff_medians = []
    diff_lower_errors = []
    diff_upper_errors = []

    for ii, sim_fn in enumerate(simulation_functions):
        n_sims = sim_fn(n_obs)
        kl_differences = []

        for i in range(num_seeds):
            # Octiles
            dir_path_octile = os.path.join(
                base_dir_octiles,
                f"npe_n_obs_{n_obs}_n_sims_{n_sims}_seed_{str(i)}"
            )
            kl_file_octile = os.path.join(dir_path_octile, "kl.txt")
            kl_value_octile = None
            if os.path.isfile(kl_file_octile):
                with open(kl_file_octile, 'r') as file:
                    try:
                        kl_value = float(file.read().strip())
                        if math.isfinite(kl_value):
                            kl_value_octile = kl_value
                    except ValueError:
                        continue

            # Duodeciles
            dir_path_duodecile = os.path.join(
                base_dir_duodeciles,
                f"npe_n_obs_{n_obs}_n_sims_{n_sims}_seed_{str(i)}"
            )
            kl_file_duodecile = os.path.join(dir_path_duodecile, "kl.txt")
            kl_value_duodecile = None
            if os.path.isfile(kl_file_duodecile):
                with open(kl_file_duodecile, 'r') as file:
                    try:
                        kl_value = float(file.read().strip())
                        if math.isfinite(kl_value):
                            kl_value_duodecile = kl_value
                    except ValueError:
                        continue

            # If both values are available, compute difference
            if kl_value_octile is not None and kl_value_duodecile is not None:
                diff = kl_value_duodecile - kl_value_octile
                kl_differences.append(diff)

        # Calculate statistics of differences
        if kl_differences:
            median_diff = np.median(kl_differences)
            lower_diff = median_diff - np.percentile(kl_differences, 5)
            upper_diff = np.percentile(kl_differences, 95) - median_diff
            diff_medians.append(median_diff)
            diff_lower_errors.append(lower_diff)
            diff_upper_errors.append(upper_diff)
        else:
            diff_medians.append(np.nan)
            diff_lower_errors.append(0)
            diff_upper_errors.append(0)

    # Convert labels to x positions
    x_positions = np.arange(len(n_sims_label))

    # Plotting
    plt.figure(figsize=(12, 8))

    # Plot differences
    plt.errorbar(
        x_positions,
        diff_medians,
        yerr=[diff_lower_errors, diff_upper_errors],
        fmt='-o',
        capsize=5,
        markersize=8,
        linewidth=2
    )

    plt.axhline(0, color='black', linestyle='--', linewidth=1)

    plt.xticks(x_positions, n_sims_label, fontsize=24)
    plt.xlabel("Choice of N", fontsize=24)
    plt.ylabel(r"Difference in KL Divergence", fontsize=24)
    plt.title(f"Difference in KL Divergence (Duodecile - Octile) (n_obs = {n_obs})", fontsize=26)
    plt.tick_params(axis='both', which='major', labelsize=24)
    plt.tight_layout()
    plt.savefig(f"plots/gnk_differences_{str(n_obs)}.pdf")
    plt.clf()
```
